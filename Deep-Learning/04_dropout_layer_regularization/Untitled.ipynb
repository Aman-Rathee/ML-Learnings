```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./sonar_dataset.csv', header=None)
df.head()
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
df[60].value_counts()
```

```python
x = df.drop(60, axis='columns')
y = df[60]
y.head()
```

```python
y = pd.get_dummies(y, drop_first=True, dtype=int)
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.25)
X_train.shape, X_test.shape
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
model = keras.Sequential([
    keras.layers.Dense(60, input_dim=60, activation='relu'),
    keras.layers.Dense(30, activation='relu'),
    keras.layers.Dense(10, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid'),
])

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

model.fit(X_train, y_train, epochs=70, batch_size=9)
```

```python
model.evaluate(X_test, y_test)
```

```python
y_pred = model.predict(X_test).reshape(-1)
print(y_pred[:10])

y_pred = np.round(y_pred)
print(y_pred[:10])
```

```python
from sklearn.metrics import confusion_matrix, classification_report
print(classification_report(y_test, y_pred))
```

```python
modelD = keras.Sequential([
    keras.layers.Dense(60, input_dim=60, activation='relu'),
    keras.layers.Dropout(0.3),
    keras.layers.Dense(30, activation='relu'),
    keras.layers.Dropout(0.3),
    keras.layers.Dense(10, activation='relu'),
    keras.layers.Dropout(0.3),
    keras.layers.Dense(1, activation='sigmoid'),
])

modelD.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

modelD.fit(X_train, y_train, epochs=70, batch_size=9)
```

```python
modelD.evaluate(X_test, y_test)
```

```python
y_pred = modelD.predict(X_test).reshape(-1)
print(y_pred[:10])

y_pred = np.round(y_pred)
print(y_pred[:10])
```

```python
print(classification_report(y_test, y_pred))
```

## You can see that by using dropout layer test accuracy increased from 0.77 to 0.81
