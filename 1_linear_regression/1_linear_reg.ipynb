```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("homeprices.csv")
df
```

```python
%matplotlib inline
plt.xlabel('area(sqr ft)')
plt.ylabel('price(US$)')
plt.scatter(df.area, df.price, color='red', marker='+')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']], df.price)
```

```python
reg.predict([[5000]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
%matplotlib inline
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area, df.price, color='red', marker='*')
plt.plot(df.area, reg.predict(df[['area']]), color='green')
```

```python
data = pd.read_csv("areas.csv")
data.head()
```

```python
p = reg.predict(data)
```

```python
p
```

```python
data['prices'] = p
```

```python
data.to_csv("prediction.csv", index=False)
```
